### using more than one filter

    filt = (df['column name']   contionational operator comparision value   value to compare with) and/or ((df['column name']   contionational operator comparision value   value to compare with))
    
    filt = (df["Country"] == 'India') and/or (df["Salary"] > 10000000)
    
    df[filt] for complete dataframe
    
    df.loc[filt,list of Columns we want info from]

```python
salary_laguege = langauge_filt & high_salary
```

```python
df.loc[salary_laguege , ['Age', 'Country','YearsCode','ConvertedComp','LanguageWorkedWith']]
```





## Negation of the filters

    filt = (df['column name']   contionational operator comparision value   value to compare with) and/or ((df['column name']   contionational operator comparision value   value to compare with))
    
    filt = (df["Country"] == 'India') and/or (df["Salary"] > 10000000)
    
    df[~filt] for complete dataframe
    
    df.loc[~filt,list of Columns we want info from]
    
### adding the tilda operator in front of the filter will give all the values other than the output of filter

```python
df.loc[~salary_laguege , ['Age', 'Country','YearsCode','ConvertedComp','LanguageWorkedWith']]
```

## apply - used for calling a function on our values it can either work on data frame or  a series object behaviour is a bit different in both cases.

    
## map 
## applymap - Only works for dataframes Series objects don't have applymap
## replace

## Apply
### On series 
    
    df['column name'].apply(function you want to apply)
    
    eg.
    1.(to check the length of our email ids)
    
    people_df['EMAIL_PERSON'].apply(len)
    2.to make ids lower case
    
    def update_email(email):
        return email.lower()
    people_df['EMAIL_PERSON'].apply(update_email)
    
    #don't put "()" we dont want execute it here just pass it as a function.
    
### On dataframe

    df.apply(function)
    
    here the function will get applied to Series in the dataframes that is either a row or a columns depending on the choice of the axis. default axis is row so all functions get applied to columns since axis of the columns is rows and vice versa for rows  


```python
def update_email(email):
        return email.lower()
people_df['EMAIL_PERSON'] = people_df['EMAIL_PERSON'].apply(update_email)
people_df
```

```python
people_df['EMAIL_PERSON'] = people_df['EMAIL_PERSON'].apply(lambda x: x.upper())
people_df
```

## Map

Works only for Series objects you cannot use it for the dataframe directly.if you want to make the changes permnent then use df[col] = df[col].map({}) 
eg.

    df[column name].map({'values you want to change' : 'the changed values to apper'})
    
    df['NAME_PERSON'].map({'Pranav':'Sunil','Pratik':'Madhuri'})
    
### it will introduce nan values in the Series for which you have not mentioned the change so keep it in mind

```python
people_df['NAME_PERSON'].map({'Pranav':'Sunil','Pratik':'Madhuri'})
```

## Replace

Works same as map does not introduce the nan values just finds and replaces what we have mentioned in the dictionary
if you want to make the changes permnent then use df[col] = df[col].map({})

```python
people_df['NAME_PERSON'].replace({'Pranav':'Sunil','Pratik':'Madhuri'})
```
